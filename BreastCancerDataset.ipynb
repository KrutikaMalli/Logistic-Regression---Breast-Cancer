## ***DERTERMINATION OF BREAST CANCER USING LOGISTIC*** ***REGRESSION***

## IMPORTING THE LIBRARIES

```python
import pandas as pd
```

## IMPORTING THE DATASET

```python
dataset = pd.read_csv('breast_cancer_dataset(Logistic_Regression).csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## SPLITTING THE DATASET INTO TRAINING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.20, random_state = 0)
```

## TRAINING THE LOGISTIC REGRESSION ON THE TRAINING SET

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression( random_state = 0)
classifier.fit(X_train, y_train)
```

## PREDICTING THE TEST SET RESULTS

```python
y_pred = classifier.predict(X_test)
```

## CREATING THE CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
# 131 correct predictions and 6 incorrect predictions.
```

## COMPUTING THE ACCURACY WITH k-FOLD CROSS VALIDATION

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %" .format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %" .format(accuracies.std()*100))
```
